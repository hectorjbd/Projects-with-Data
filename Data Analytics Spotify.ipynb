## Importando Librerias

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

## Importando Dataframe

```python
df_tracks = pd.read_csv('C:\\Users\\ADMIN\\Documents\\Cursos\\Data Projects\\Project 2 - Spotify\\tracks.csv')
df_tracks.head()
```

## Canciones con popularidad mayor a 90

```python
mas_popular = df_tracks.query('popularity>90', inplace = False).sort_values('popularity', ascending = False)
mas_popular[:10]
```

#### Colocamos como indice del dataframe, el año de lanzamiento

```python
df_tracks.set_index('release_date', inplace = True)
df_tracks.index = pd.to_datetime(df_tracks.index)
df_tracks.head()
```

#### La duracion de las canciones se encuentran en milisegundos, la transformaremos a segundos

```python
df_tracks['duration'] = df_tracks['duration_ms'].apply(lambda x: round(x/1000))
df_tracks.drop('duration_ms', inplace = True , axis = 1)
```

```python
df_tracks.duration.head()
```

## Buscaremos si existe correlacion con un mapa de calor

```python
corr_df = df_tracks.drop(["key","mode","explicit"],axis = 1).corr(method = "pearson")
plt.figure(figsize=(14,6))
heatmap = sns.heatmap(corr_df,annot = True, fmt =".1g" , vmin=-1, vmax=1, center=0, cmap="inferno", linewidths=1, linecolor="Black")
heatmap.set_title("Correlacion Mapa de Calor entre variables")
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation = 90)
```

#### Podemos apreciar la correlacion entre las distintas variables. Por ejemplo en la columna de "Energy" si la interceptamos con la fila "Loudness" podemos observar, que existe una correlacion de 0.8, eso quiere decir que tienen una alta correlacion entre si. Mientras que para la columna de "Acousticness" y para la fila de "Energy", se observa una correlacion negativa de -0.7, lo que significa que estas dos variables son inversamente proporcionales.

## Realizaremos una correlacion entre "Energy" y "Loudness"

```python
sample_df = df_tracks.sample(int(0.004*len(df_tracks)))
```

```python
print(len(sample_df))
```

```python
plt.figure(figsize=(10,6))
```

```python
sns.regplot(data = sample_df, y = "loudness" , x = "energy", color = "c").set(title = "Correlacion Volumen vs Energia")
```

#### Observamos que existe una correlacion positiva entre energia y volumen

## Realizaremos una correlacion negativa entre "Energy" y "Acousticness"

```python
plt.figure(figsize=(10,6))
sns.regplot(data = sample_df, y = "energy" , x = "acousticness", color = "b").set(title = "Correlacion Acustica vs Energia")
```

#### Observamos que existe una correlacion negativa entre energia y acustica

```python
df_tracks['dates'] = df_tracks.index.get_level_values('release_date')
df_tracks.dates = pd.to_datetime(df_tracks.dates)
years = df_tracks.dates.dt.year
```

## Numero de canciones por año

```python
sns.displot(years, discrete = True, aspect = 2, height = 5, kind = 'hist').set(title = "Numero de canciones por año")
```

## Año vs Duracion de canciones. Diagrama de barra

```python
total_dr = df_tracks.duration
fig_dim = (18,9)
fig, ax = plt.subplots(figsize = fig_dim)
fig = sns.barplot(x = years, y = total_dr, ax = ax, errwidth= False).set(title = "Año vs Duracion (segundos)")
plt.xticks(rotation = 90)
```
